```python
# ============================================================================
# STEP 1: IMPORT LIBRARIES AND SETUP
# ============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                           f1_score, confusion_matrix, classification_report)
import xgboost as xgb
from xgboost import XGBClassifier
import warnings
warnings.filterwarnings('ignore')

# Set style for better plots
plt.style.use('default')
sns.set_palette("husl")

print("‚úÖ Step 1 Complete: All libraries imported successfully!")
```

```python

# ============================================================================
# STEP 2: LOAD THE DATASETS
# ============================================================================

# Load the three CSV files
print("üìÇ Loading datasets...")

try:
    train_df = pd.read_csv("/content/drive/MyDrive/UpdatedDataset/training.csv")
    val_df = pd.read_csv("/content/drive/MyDrive/UpdatedDataset/validating.csv")
    test_df = pd.read_csv("/content/drive/MyDrive/UpdatedDataset/testing.csv")

    print(f"‚úÖ Training data loaded: {train_df.shape}")
    print(f"‚úÖ Validation data loaded: {val_df.shape}")
    print(f"‚úÖ Test data loaded: {test_df.shape}")

    # Display basic info about the data
    print("\nüìä Dataset Information:")
    print(f"Training set columns: {len(train_df.columns)}")
    print(f"Sample of column names: {list(train_df.columns[:5])}")
    print(f"Production range in training: {train_df['Production (M.Ton)'].min():.2f} - {train_df['Production (M.Ton)'].max():.2f}")

except FileNotFoundError as e:
    print("‚ùå Error: CSV files not found. Please ensure train.csv, validation.csv, and test.csv are in the current directory.")
    print("Current working directory files should include your CSV files.")

print("\n‚úÖ Step 2 Complete: Data loading finished!")
```

```python
# ============================================================================
# STEP 3: CREATE PRODUCTION CATEGORIES (5 LEVELS)
# ============================================================================

def create_production_categories(df, thresholds=None):
    """Create 5 production level categories"""
    df = df.copy()
    production_col = 'Production (M.Ton)'

    # Calculate thresholds based on training data quantiles
    if thresholds is None:
        thresholds = [
            df[production_col].quantile(0.2),   # Very Low: 0-20%
            df[production_col].quantile(0.4),   # Low: 20-40%
            df[production_col].quantile(0.6),   # Medium: 40-60%
            df[production_col].quantile(0.8),   # High: 60-80%
            df[production_col].max()            # Very High: 80-100%
        ]

    # Create categories
    conditions = [
        df[production_col] <= thresholds[0],
        (df[production_col] > thresholds[0]) & (df[production_col] <= thresholds[1]),
        (df[production_col] > thresholds[1]) & (df[production_col] <= thresholds[2]),
        (df[production_col] > thresholds[2]) & (df[production_col] <= thresholds[3]),
        df[production_col] > thresholds[3]
    ]

    class_names = ['Very Low', 'Low', 'Medium', 'High', 'Very High']
    df['Production_Category'] = np.select(conditions, class_names, default='Medium')

    return df, thresholds

# Apply categorization
print("üè∑Ô∏è Creating production categories...")

# Create categories using training data thresholds
train_df_cat, production_thresholds = create_production_categories(train_df)
val_df_cat, _ = create_production_categories(val_df, production_thresholds)
test_df_cat, _ = create_production_categories(test_df, production_thresholds)

# Display category distribution
print("\nüìä Production Category Distribution:")
print("Training set:")
print(train_df_cat['Production_Category'].value_counts())
print("\nValidation set:")
print(val_df_cat['Production_Category'].value_counts())
print("\nTest set:")
print(test_df_cat['Production_Category'].value_counts())

print(f"\nüìè Production thresholds used:")
class_names = ['Very Low', 'Low', 'Medium', 'High', 'Very High']
for i, (name, threshold) in enumerate(zip(class_names, production_thresholds)):
    if i == 0:
        print(f"{name}: ‚â§ {threshold:.2f}")
    else:
        print(f"{name}: {production_thresholds[i-1]:.2f} < x ‚â§ {threshold:.2f}")

print("\n‚úÖ Step 3 Complete: Production categories created!")
```

```python
# ============================================================================
# STEP 4: PREPARE FEATURES AND TARGETS
# ============================================================================

def prepare_features_targets(df):
    """Prepare features and target variables"""
    # Select feature columns (exclude production columns)
    feature_cols = [col for col in df.columns if col not in ['Production (M.Ton)', 'Production_Category']]

    X = df[feature_cols].copy()
    y = df['Production_Category'].copy()

    # Handle missing values by filling with mean
    X_filled = X.fillna(X.mean())

    return X_filled, y, feature_cols

print("üîß Preparing features and targets...")

# Prepare data for all sets
X_train, y_train, feature_names = prepare_features_targets(train_df_cat)
X_val, y_val, _ = prepare_features_targets(val_df_cat)
X_test, y_test, _ = prepare_features_targets(test_df_cat)

# Encode target labels
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_val_encoded = label_encoder.transform(y_val)
y_test_encoded = label_encoder.transform(y_test)

print(f"‚úÖ Features prepared: {X_train.shape[1]} features")
print(f"‚úÖ Feature names stored: {len(feature_names)} features")
print(f"‚úÖ Labels encoded: {len(label_encoder.classes_)} classes")
print(f"Class mapping: {dict(zip(label_encoder.classes_, range(len(label_encoder.classes_))))}")

print("\n‚úÖ Step 4 Complete: Features and targets prepared!")
```

```python
# ============================================================================
# STEP 5: TRAIN XGBOOST MODEL WITH REAL-TIME PROGRESS DISPLAY
# ============================================================================

def train_xgboost_with_realtime_display(X_train, y_train, X_val, y_val, n_epochs=250):
    """Train XGBoost model with real-time progress display"""

    print("üöÄ Training model with real-time progress display...")
    print("Training model with history tracking...")

    # Initialize training history storage
    training_history = {
        'epochs': [],
        'train_loss': [],
        'val_loss': [],
        'train_acc': [],
        'val_acc': []
    }

    # Train model iteratively to show progress
    model = None
    best_val_loss = float('inf')
    patience_counter = 0
    early_stopping_rounds = 50

    for epoch in range(1, n_epochs + 1):
        # Create model with current number of estimators
        current_model = XGBClassifier(
            n_estimators=epoch,
            max_depth=6,
            learning_rate=0.1,
            subsample=0.8,
            colsample_bytree=0.8,
            random_state=42,
            eval_metric='mlogloss'
        )

        # Train model
        eval_set = [(X_train, y_train), (X_val, y_val)]
        current_model.fit(
            X_train, y_train,
            eval_set=eval_set,
            verbose=False
        )

        # Make predictions
        train_pred = current_model.predict(X_train)
        val_pred = current_model.predict(X_val)

        # Calculate accuracies
        train_acc = accuracy_score(y_train, train_pred)
        val_acc = accuracy_score(y_val, val_pred)

        # Get losses from evaluation results
        train_loss = current_model.evals_result_['validation_0']['mlogloss'][-1]
        val_loss = current_model.evals_result_['validation_1']['mlogloss'][-1]

        # Store history
        training_history['epochs'].append(epoch)
        training_history['train_loss'].append(train_loss)
        training_history['val_loss'].append(val_loss)
        training_history['train_acc'].append(train_acc)
        training_history['val_acc'].append(val_acc)

        # Display progress
        print(f"Epoch {epoch:3d}/{n_epochs} - "
              f"Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}, "
              f"Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}")

        # Early stopping logic
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            model = current_model  # Save best model
            patience_counter = 0
        else:
            patience_counter += 1

        if patience_counter >= early_stopping_rounds:
            print(f"Early stopping at epoch {epoch}")
            break

    print("Training completed.")

    # Store best iteration info in training history for reference
    training_history['best_iteration'] = len(training_history['epochs'])

    return model, training_history

# Train the model with real-time display
model, training_history = train_xgboost_with_realtime_display(
    X_train.values, y_train_encoded,
    X_val.values, y_val_encoded,
    n_epochs=250
)

print(f"\n‚úÖ Training complete! Best iteration: {training_history.get('best_iteration', 'N/A')}")
print("\n‚úÖ Step 5 Complete: Model training with real-time display finished!")
```

```python
# ============================================================================
# STEP 6: PLOT TRAINING PROGRESS ANALYSIS
# ============================================================================

def plot_training_progress_analysis(training_history):
    """Plot training progress with loss and accuracy curves"""

    # Set up the figure with subplots
    fig = plt.figure(figsize=(16, 6))

    # Add main title
    fig.suptitle('Training Progress Analysis', fontsize=16, fontweight='bold', y=0.98)

    # Create subplots
    ax1 = plt.subplot(1, 2, 1)  # Loss plot
    ax2 = plt.subplot(1, 2, 2)  # Accuracy plot

    # Plot 1: Model Loss Over Training
    ax1.plot(training_history['epochs'], training_history['train_loss'],
             'b-', linewidth=2.5, label='Training Loss')
    ax1.plot(training_history['epochs'], training_history['val_loss'],
             'r-', linewidth=2.5, label='Validation Loss')

    ax1.set_title('Model Loss Over Training', fontsize=14, fontweight='bold', pad=15)
    ax1.set_xlabel('Iteration', fontsize=12)
    ax1.set_ylabel('Multi-class Log Loss', fontsize=12)
    ax1.legend(fontsize=11, loc='upper right')
    ax1.grid(True, alpha=0.3)
    ax1.set_xlim(0, max(training_history['epochs']))

    # Plot 2: Model Accuracy Over Training
    ax2.plot(training_history['epochs'], training_history['train_acc'],
             'b-', linewidth=2.5, label='Training Accuracy')
    ax2.plot(training_history['epochs'], training_history['val_acc'],
             'orange', linewidth=2.5, label='Validation Accuracy')

    ax2.set_title('Model Accuracy Over Training', fontsize=14, fontweight='bold', pad=15)
    ax2.set_xlabel('Iteration', fontsize=12)
    ax2.set_ylabel('Accuracy', fontsize=12)
    ax2.legend(fontsize=11, loc='lower right')
    ax2.grid(True, alpha=0.3)
    ax2.set_xlim(0, max(training_history['epochs']))
    ax2.set_ylim(0, 1)

    # Adjust layout
    plt.tight_layout()
    plt.subplots_adjust(top=0.9)
    plt.show()

    # Print training progress summary
    print(f"\nüìä Training Progress Summary:")
    print(f"- Total epochs trained: {len(training_history['epochs'])}")
    print(f"- Final training loss: {training_history['train_loss'][-1]:.4f}")
    print(f"- Final validation loss: {training_history['val_loss'][-1]:.4f}")
    print(f"- Final training accuracy: {training_history['train_acc'][-1]:.4f}")
    print(f"- Final validation accuracy: {training_history['val_acc'][-1]:.4f}")
    print(f"- Best iteration (early stopping): {training_history.get('best_iteration', 'N/A')}")

print("üìà Generating training progress visualizations...")
plot_training_progress_analysis(training_history)

print("\n‚úÖ Step 6 Complete: Training progress analysis plotted!")
```

```python
# ============================================================================
# STEP 7: EVALUATE MODEL AND CALCULATE ALL METRICS
# ============================================================================

def calculate_all_metrics(model, X, y, dataset_name):
    """Calculate comprehensive metrics"""
    y_pred = model.predict(X)

    # Calculate all required metrics
    accuracy = accuracy_score(y, y_pred)
    precision = precision_score(y, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y, y_pred, average='weighted', zero_division=0)
    f1_weighted = f1_score(y, y_pred, average='weighted', zero_division=0)
    f1_macro = f1_score(y, y_pred, average='macro', zero_division=0)

    metrics = {
        'Dataset': dataset_name,
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score (Weighted Avg)': f1_weighted,
        'Macro F1 Score': f1_macro
    }

    return metrics, y_pred

print("üìä Evaluating model on all datasets...")

# Evaluate on training set
train_metrics, train_pred = calculate_all_metrics(model, X_train.values, y_train_encoded, "Training")

# Evaluate on validation set
val_metrics, val_pred = calculate_all_metrics(model, X_val.values, y_val_encoded, "Validation")

# Evaluate on test set
test_metrics, test_pred = calculate_all_metrics(model, X_test.values, y_test_encoded, "Test")

# Create results summary
results_df = pd.DataFrame([train_metrics, val_metrics, test_metrics])

print("\nüéØ MODEL EVALUATION RESULTS:")
print("="*80)
print(results_df.round(4).to_string(index=False))
print("="*80)

print("\n‚úÖ Step 7 Complete: Model evaluation finished!")
```

```python
# ============================================================================
# STEP 8: PLOT CONFUSION MATRICES
# ============================================================================

def plot_confusion_matrix(y_true, y_pred, dataset_name, class_names):
    """Plot confusion matrix"""
    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
               xticklabels=class_names,
               yticklabels=class_names,
               cbar_kws={'label': 'Number of Samples'})

    plt.title(f'Confusion Matrix - {dataset_name}', fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('Predicted Label', fontsize=12, fontweight='bold')
    plt.ylabel('True Label', fontsize=12, fontweight='bold')

    # Add accuracy to the plot
    accuracy = accuracy_score(y_true, y_pred)
    plt.text(0.02, 0.98, f'Accuracy: {accuracy:.3f}',
             transform=plt.gca().transAxes, fontsize=8,
             verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

    plt.tight_layout()
    plt.show()

print("üìà Plotting confusion matrices...")

class_names = ['Very Low', 'Low', 'Medium', 'High', 'Very High']

# Plot confusion matrix for validation set
plot_confusion_matrix(y_val_encoded, val_pred, "Validation Set", class_names)

# Plot confusion matrix for test set
plot_confusion_matrix(y_test_encoded, test_pred, "Test Set", class_names)

print("\n‚úÖ Step 8 Complete: Confusion matrices plotted!")
```

```python
# ============================================================================
# STEP 9: DETAILED CLASSIFICATION REPORTS
# ============================================================================

print("üìã Generating detailed classification reports...")

print("\n" + "="*60)
print("DETAILED CLASSIFICATION REPORT - VALIDATION SET")
print("="*60)
val_report = classification_report(
    y_val_encoded, val_pred,
    target_names=class_names,
    zero_division=0
)
print(val_report)

print("\n" + "="*60)
print("DETAILED CLASSIFICATION REPORT - TEST SET")
print("="*60)
test_report = classification_report(
    y_test_encoded, test_pred,
    target_names=class_names,
    zero_division=0
)
print(test_report)

print("\n‚úÖ Step 9 Complete: Classification reports generated!")
```

```python


# ============================================================================
# STEP 10: FEATURE IMPORTANCE ANALYSIS
# ============================================================================

def plot_feature_importance(model, feature_names, top_n=20):
    """Plot top N most important features"""

    # Get feature importance
    importance = model.feature_importances_

    # Get top N features
    indices = np.argsort(importance)[::-1][:top_n]
    top_features = [feature_names[i] for i in indices]
    top_importance = importance[indices]

    # Create plot
    plt.figure(figsize=(12, 8))
    colors = plt.cm.viridis(np.linspace(0, 1, len(top_importance)))

    bars = plt.barh(range(len(top_importance)), top_importance[::-1], color=colors[::-1])
    plt.yticks(range(len(top_importance)), top_features[::-1])
    plt.xlabel('Importance Score', fontsize=12, fontweight='bold')
    plt.title(f'Top {top_n} Most Important Features', fontsize=16, fontweight='bold', pad=20)

    # Add value labels on bars
    for i, (bar, val) in enumerate(zip(bars, top_importance[::-1])):
        plt.text(val + 0.001, bar.get_y() + bar.get_height()/2,
                f'{val:.3f}', ha='left', va='center', fontsize=9)

    plt.tight_layout()
    plt.show()

    # Print top 10 features
    print(f"\nüèÜ Top 10 Most Important Features:")
    for i, (feature, score) in enumerate(zip(top_features[:10], top_importance[:10])):
        print(f"{i+1:2d}. {feature:<40} {score:.4f}")

print("üìä Analyzing feature importance...")
plot_feature_importance(model, feature_names, top_n=20)

print("\n‚úÖ Step 10 Complete: Feature importance analysis finished!")
```
